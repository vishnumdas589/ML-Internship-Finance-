### DIVERSIFICATION AND CHECKING CORRELATION OF MULTIPLE STOCKS

### CALCULATING VOLATITILITY  OF THE STOCKS AND SETTING BETA MODEL

```python
# Adding volititly column for each stock
per_change['nify voli'] = per_change.NIFY.rolling(9).std()
per_change['reliance voli']= per_change.RELIANCE.rolling(9).std()
per_change['axis voli'] = per_change.AXIS.rolling(9).std()
per_change['tcs voli'] = per_change.TCS.rolling(9).std()
per_change['asian voli'] = per_change.ASIAN.rolling(9).std()
```

```python
# Viewing the new columns
per_change.head(10)
```

```python
# Plotting the Beta indicator for Reliance and Axis bank
fig,ax= plt.subplots(figsize=(15,8))
ax.plot(per_change['nify voli'])
ax.plot(per_change['reliance voli'])
ax.plot(per_change['axis voli'])
# Adding legend 
ax.legend(per_change.columns[:3],title = 'VOLATILITY')
```

```python
# Plotting beta for TCS and Asian Paints
fig,ax= plt.subplots(figsize=(15,8))
ax.plot(per_change['nify voli'],color='black')
ax.plot(per_change['tcs voli'])
ax.plot(per_change['asian voli'])
ax.legend(per_change.columns[[0,3,4]],title = 'VOLATILITY')
```

### TRADE CALLS - USING SIMPLE MOVING AVERAGE 

```python
# Creating new DateFrame containing just the closing price and date of nifty
nifty =pd.DataFrame()
nifty['DATE'] = closing_price.Date
nifty['NIFY']= closing_price['NIFY']
```

```python
# Creating the 21 and 34 days moving average 
nifty['SMA(21)'] = closing_price['NIFY'].rolling(21).mean()
nifty['SMA(34)'] = closing_price['NIFY'].rolling(34).mean()
nifty.index = nifty.DATE
nifty.index = pd.to_datetime(nifty.index)
nifty = nifty.drop(columns='DATE')
nifty
```

```python
# Plotting the nify closing price along with 21 amd 34 day moving averages 
fig,ax = plt.subplots(figsize=(15,5));
ax.plot(nifty['NIFY'],color='black')
ax.plot(nifty['SMA(21)'])
ax.plot(nifty['SMA(34)'])
ax.legend(nifty.columns[:])
```

```python
# identifying buy/sell zone
# nifty['SMA(21)']= nifty['SMA(21)'].fillna(0)
# nifty['SMA(34)']= nifty['SMA(34)'].fillna(0)
nifty['BUY']= np.where((nifty['SMA(21)']>nifty['SMA(34)']),1,0)
nifty['SELL'] =  np.where((nifty['SMA(21)']<nifty['SMA(34)']),1,0)
nifty
```

```python
# identifying buy/sell signal
nifty['Buy_Ind']=np.where((nifty["BUY"]>nifty['BUY'].shift(1)),1,0)
nifty['Sell_Ind']=np.where((nifty["SELL"]>nifty['SELL'].shift(1)),1,0)
nifty
```

```python
# ploting the graph with buy/sell signal
fig,ax = plt.subplots(figsize=(15,10));
ax.plot(nifty['NIFY']);
ax.plot(nifty['SMA(21)']);
ax.plot(nifty['SMA(34)']);
ax.legend(nifty.columns[:]);
ax.scatter(nifty[nifty['Buy_Ind']==1].index.values,nifty[nifty['Buy_Ind']==1].NIFY.values,label='skitscat',color ='green',s=70,marker='^')
ax.scatter(nifty[nifty['Sell_Ind']==1].index.values,nifty[nifty['Sell_Ind']==1].NIFY.values,label='skitscat',color= 'red',s=70,marker='v')
```

### TRADE CALLS - USING BOLLINGER BANDS

```python
# Creating the 14 day average column
nifty['AVG(14)'] = nifty.NIFY.rolling(14).mean()
nifty
```

```python
# Creating the upper and lower columns with values 2 times standard deviation above and bleow the closing price 
nifty['Upper']= nifty['AVG(14)'] + 2*(nifty['NIFY'].rolling(14).std())
nifty['Lower']= nifty['AVG(14)'] - 2*(nifty['NIFY'].rolling(14).std())
nifty
```

```python
# Plotting the Bollinger Band
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(nifty['NIFY'],color='black')
ax.plot(nifty['AVG(14)'])
ax.plot(nifty['Upper'])
ax.plot(nifty['Lower'])
```
